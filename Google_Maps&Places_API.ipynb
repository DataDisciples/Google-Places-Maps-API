```python
import googlemaps
import time
import csv
from math import radians, sin, cos, sqrt, atan2, ceil
import datetime


api_key = 'AIzaSyDujvnig0KQ0SLP_P9DTDNfRlrsIfQnFQs'
gmaps = googlemaps.Client(key=api_key)


def get_place_details(place_id):
    try:
        return gmaps.place(place_id=place_id, fields=[
            'business_status', 'formatted_phone_number', 'geometry/location', 'icon',
            'international_phone_number', 'name', 'opening_hours', 'photo', 'place_id',
            'plus_code', 'price_level', 'rating', 'review', 'reviews', 'url',
            'user_ratings_total', 'utc_offset', 'vicinity', 'website', 'address_component',
            'formatted_address'
        ])
    except googlemaps.exceptions.ApiError as e:
        print(f"API error fetching details for place_id {place_id}: {e}")
        time.sleep(60)
        return get_place_details(place_id)
    except Exception as e:
        print(f"Error fetching details for place_id {place_id}: {e}")
        return None


def clean_review_text(text):
    return text.replace('\n', ' ').replace('\r', '').replace('"', '""')[:100] + '...' if text else ''


def haversine_distance(lat1, lon1, lat2, lon2):
    R = 3959
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    return R * c


marion_lat, marion_lon = 37.7306, -88.9331


def search_businesses(center_lat, center_lon, radius_miles, csvwriter, seen_place_ids):
    radius_meters = radius_miles * 1609
    results_count = 0
    next_page_token = None

    while True:
        try:
            results = gmaps.places_nearby(location=(center_lat, center_lon), radius=radius_meters, page_token=next_page_token)
        except googlemaps.exceptions.ApiError as e:
            print(f"API error occurred: {e}. Waiting for 60 seconds before retrying.")
            time.sleep(60)
            continue
        except Exception as e:
            print(f"Error occurred: {e}. Waiting for 60 seconds before retrying.")
            time.sleep(60)
            continue

        for result in results.get('results', []):
            place_id = result['place_id']
            if place_id in seen_place_ids:
                continue
            seen_place_ids.add(place_id)

            details = get_place_details(place_id)
            if not details:
                continue

            details = details.get('result', {})
            name = details.get('name', 'N/A')
            address = details.get('formatted_address', 'N/A')
            phone = details.get('formatted_phone_number', 'N/A')
            website = details.get('website', 'N/A')
            rating = details.get('rating', 'N/A')
            types = ', '.join(details.get('types', ['N/A'])) if 'types' in details else 'N/A'
            business_status = details.get('business_status', 'N/A')
            international_phone_number = details.get('international_phone_number', 'N/A')
            url = details.get('url', 'N/A')
            user_ratings_total = details.get('user_ratings_total', 'N/A')
            utc_offset = details.get('utc_offset', 'N/A')
            vicinity = details.get('vicinity', 'N/A')
            price_level = details.get('price_level', 'N/A')

            lat = result['geometry']['location']['lat']
            lon = result['geometry']['location']['lng']
            distance = haversine_distance(marion_lat, marion_lon, lat, lon)

            if distance <= 25:
                reviews = details.get('reviews', [])
                review_texts = [clean_review_text(review.get('text', '')) for review in reviews[:3]]
                review_texts += ['N/A'] * (3 - len(review_texts))

                csvwriter.writerow([
                    name, address, phone, website, rating, types, f"{distance:.2f}", business_status,
                    international_phone_number, url, user_ratings_total, utc_offset, vicinity, price_level
                ] + review_texts)
                results_count += 1

            time.sleep(1)

        next_page_token = results.get('next_page_token')
        if not next_page_token:
            break

        time.sleep(2)

    return results_count


start_time = datetime.datetime.now()
filename = f'businesses_25miles_from_marion_{start_time.strftime("%Y%m%d_%H%M%S")}.csv'

with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile, quoting=csv.QUOTE_ALL)
    csvwriter.writerow([
        'Name', 'Address', 'Phone', 'Website', 'Rating', 'Types', 'Distance (miles)',
        'Business Status', 'International Phone Number', 'URL', 'User Ratings Total',
        'UTC Offset', 'Vicinity', 'Price Level', 'Review 1', 'Review 2', 'Review 3'
    ])

    total_results = 0
    seen_place_ids = set()


    grid_size = 3
    num_points = ceil(50 / grid_size)

    for i in range(-num_points, num_points + 1):
        for j in range(-num_points, num_points + 1):
            lat = marion_lat + (i * grid_size / 69)
            lon = marion_lon + (j * grid_size / (69 * cos(radians(marion_lat))))
            if haversine_distance(marion_lat, marion_lon, lat, lon) <= 25:
                print(f"Searching at coordinates: {lat:.4f}, {lon:.4f}")
                results = search_businesses(lat, lon, grid_size, csvwriter, seen_place_ids)
                total_results += results
                print(f"Found {results} new businesses. Total so far: {total_results}")


                if total_results % 500 == 0:
                    print(f"Reached {total_results} results. Saving progress...")
                    csvfile.flush()


                time.sleep(2)

end_time = datetime.datetime.now()
duration = end_time - start_time

print(f"Total businesses written to CSV: {total_results}")
print(f"Data has been saved to '{filename}'")
print(f"Total runtime: {duration}")
```
